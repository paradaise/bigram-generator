```python
import random

# Пример текста
text = open("pushkin.txt", encoding = "utf-8").read()

# Разделение текста на биграммы
words = text.split()
bigram_list = [(words[i], words[i+1]) for i in range(len(words)-1)]

print(f"биграмы-"bigram_list)


fdist = {}
for bigram in bigram_list:
    if bigram in fdist:
        fdist[bigram] += 1
    else:
        fdist[bigram] = 1

# Генерация текста на основе биграмм
generated_text = []
current_bigram = random.choice(bigram_list)
generated_text.extend(list(current_bigram))

for i in range(10):  # Генерация 10 биграмм
    next_word_options = [key[1] for key in fdist if key[0] == current_bigram[1]]
    next_word = random.choice(next_word_options)
    generated_text.append(next_word)
    current_bigram = (current_bigram[1], next_word)

result_text = ' '.join(generated_text)
print(result_text)
```
